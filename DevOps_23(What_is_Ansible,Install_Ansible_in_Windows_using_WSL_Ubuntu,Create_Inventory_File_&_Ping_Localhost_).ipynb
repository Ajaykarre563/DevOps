# **Install Ansible in Windows using WSL Ubuntu**

# **✅ Step 1: Enable WSL in Windows**
Open PowerShell as Administrator and run:



```
 wsl --install
```



# **✅ Step 2: Install Ubuntu via Microsoft Store**
Go to the Microsoft Store.

Search for "Ubuntu" (choose version like Ubuntu 22.04).

Click Install and wait for setup.

Launch Ubuntu → set a new UNIX username and password.
