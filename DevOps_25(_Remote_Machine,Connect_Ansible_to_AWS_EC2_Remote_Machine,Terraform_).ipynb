**In Ansible, the local machine is the computer where Ansible is installed (also called the control node). The remote machine (also called a managed node) is any computer or server that Ansible connects to in order to run commands or deploy applications.**

Local Machine: Where you write and run Ansible playbooks (e.g., your laptop or desktop).

Remote Machine: The target server where you want Ansible to perform tasks (e.g., install software, update config files).

# **Terraform**

**Terraform is an open-source Infrastructure as Code (IaC) tool developed by HashiCorp.**

**It lets you automate the creation, update, and management of cloud resources like servers, databases, networks, etc.**

# **‚úÖ Key Points:**

You write code to define infrastructure (like AWS EC2, S3, Azure VMs, etc.).

Code is written in HCL (HashiCorp Configuration Language).

Supports multiple cloud providers (AWS, Azure, GCP, etc.).

Helps you version control and reproduce infrastructure easily.
