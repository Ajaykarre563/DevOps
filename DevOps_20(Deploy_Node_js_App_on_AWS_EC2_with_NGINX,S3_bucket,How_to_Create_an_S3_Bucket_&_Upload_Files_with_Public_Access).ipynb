# **How to Create an S3 Bucket & Upload Files with Public Access**

# **✅ Step 1: Create an S3 Bucket**
Sign in to AWS Console.

Go to S3 under "Storage".

Click “Create bucket”.

Enter a unique Bucket name (e.g., my-public-images-bucket).

Choose AWS Region (e.g., Asia Pacific (Mumbai)).

Uncheck: “Block all public access” under Permissions.

Acknowledge the warning about public access.

Click Create bucket.

# **✅ Step 2: Upload Files to the Bucket**
Open the bucket.

Click “Upload”.

Add your file (e.g., logo.png).

Click Next until the “Set permissions” step.

Choose:

✅ Grant public read access to this object

Click Upload.
