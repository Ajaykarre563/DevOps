# **How schedule GitHub Actions workflows**

**To schedule GitHub Actions workflows, you use the schedule event with cron syntax in your workflow YAML file. This allows your workflow to run automatically at specific times (e.g., daily, weekly).**

image.png

# **⏰ Cron Format**



```
┌───────────── minute (0 - 59)
│ ┌───────────── hour (0 - 23)
│ │ ┌───────────── day of the month (1 - 31)
│ │ │ ┌───────────── month (1 - 12)
│ │ │ │ ┌───────────── day of the week (0 - 7) (Sunday = 0 or 7)
│ │ │ │ │
│ │ │ │ │
* * * * *

```



image.png

# **📝 Example Workflow**



```
name: Daily CI

on:
  schedule:
    - cron: '0 0 * * *'  # Runs daily at midnight UTC

jobs:
  build:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - name: Run a script
        run: echo "Running daily job"

```



# **⚠️ Notes**

All times in cron are in UTC.

GitHub may delay jobs to balance loads. Don’t use for precise timing needs.

# **How to Disable GitHub Actions Workflows**



```
name: Disable Workflow Video

on:
  push:
    branches: [ main ]

jobs:
  disable:
    runs-on: ubuntu-latest

    steps:
      - name: Farzeen
        run: echo "Disable Workflow Video DevOps"

```



image.png

# **⚠️ Important Note:**
If a workflow is disabled in GitHub, any changes you make to the .yml file won’t be applied (even if you push new commits) until you re-enable it.

# **automate the Angular build process**

**To automate the Angular build process, you can use GitHub Actions to automatically build your Angular app whenever you push code.**

# **🛠️ 2. Add the Workflow Code**



```
name: Angular Build

on:
  push:
    branches:
      - main  # or your main development branch
  pull_request:

jobs:
  build:
    runs-on: ubuntu-latest

    steps:
      - name: Checkout Code
        uses: actions/checkout@v4

      - name: Set up Node.js
        uses: actions/setup-node@v4
        with:
          node-version: '18'  # match your Angular version

      - name: Install Dependencies
        run: npm install

      - name: Build Angular App
        run: npm run build -- --configuration=production

```
