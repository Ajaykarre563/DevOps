# **🌐 Public Cloud**
What it is: Shared services offered by companies like AWS, Azure, or Google Cloud.

Example: You store your photos on Google Drive — you're using the public cloud. Google owns the infrastructure and you're one of many users.

# **🏢 Private Cloud**
What it is: A cloud system used by one company only, not shared with others.

Example: A bank builds its own private cloud to store customer data securely, so only it has access and full control.

# **☁️ Hybrid Cloud**
What it is: A mix of public + private cloud, used together for flexibility.

Example: An e-commerce company keeps customer payment data in a private cloud (for security), but uses the public cloud to host its website (for scalability).
