# **How to Upload Image on Docker Hub**(it us mostly used for local setup but it has a limit time for 4hrs)

# ** 1. Create an Account on Docker Hub**
Go to https://hub.docker.com/ and sign up or log in.


# ** 2. Create a Repository on Docker Hub**
Login to Docker Hub

Click "Repositories" > "Create Repository"

Name your repository (e.g., node-app)

Select Public or Private (Public is free)

Click "Create"



# **3.enter the command in vs code**



```
docker build -t ajaykarre56/my-node-app .
```



Note:



```
-t ajaykarre56/my-node-app (these code is get by repository in docker hub)

```



# ** 4. Push the Image to Docker Hub**

```
docker push ajaykarre56/my-node-app
```



# ** 5. Confirm Upload**
Go to your Docker Hub account and check the Repositories section to see your image.
