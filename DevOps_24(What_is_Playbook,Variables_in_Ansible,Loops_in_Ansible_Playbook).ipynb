**An Ansible Playbook is a YAML file that defines a set of tasks to automate configuration, deployment, and management of systems.**

# **âœ… Simple Definition:**

A playbook tells Ansible what to do and on which machines.

# **step:2 nano create_file.yml**



```
- name: Create a file using environment path
  hosts: localhost
  gather_facts: yes  # Required to get ansible_env variables
  tasks:
    - name: Create ansible_file.txt in HOME directory
      file:
        path: "{{ ansible_env.HOME }}/ansible_file.txt"
        state: touch
        mode: '0644'

```



# **âœ… 3. Run the Playbook**



```
ansible-playbook -i hosts create_file.yml

```

# **âœ… 1. Variables in Ansible**
Definition: Variables let you store reusable values like filenames, usernames, package names, etc.

# **ðŸ“˜ Example:**



```
- name: Use variables in a playbook
  hosts: localhost
  vars:
    my_file: "example.txt"
    my_dir: "/tmp"
  tasks:
    - name: Create a file using variable
      file:
        path: "{{ my_dir }}/{{ my_file }}"
        state: touch

```



my_file and my_dir are variables

{{ variable_name }} is how you use them

# **âœ… 2. Loops in Ansible**
Definition: Loops let you repeat a task multiple times with different values.



```
- name: Create multiple files
  hosts: localhost
  tasks:
    - name: Create files with loop
      file:
        path: "/tmp/{{ item }}"
        state: touch
      loop:
        - file1.txt
        - file2.txt
        - file3.txt

```



# **Variables_loops.yml**



```
- name: Playbook using variables and loops
  hosts: local
  gather_facts: true

  vars:
    file_path: "{{ ansible_env.HOME }}/files"
    file_names:
      - file1.txt
      - file2.txt
      - file3.txt

  tasks:
    - name: Ensure Directory Exists
      file:
        path: "{{ file_path }}"
        state: directory
        mode: '0755'

    - name: Create Multiple Files using Loops
      file:
        path: "{{ file_path }}/{{ item }}"
        state: touch
        mode: '0644'
      loop: "{{ file_names }}"

```



# **âœ… This playbook:**

Creates a directory named files inside the current user's home directory.

Then creates 3 text files (file1.txt, file2.txt, file3.txt) inside it.
