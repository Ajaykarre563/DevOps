# **GitLab Server with Docker Compose**

# **✅ 1. Create a docker-compose.yml File**



```
version: '3.8'
services:
  gitlab-server:
    image: 'gitlab/gitlab-ce'
    container_name: my-gitlab-server
    ports:
      - '3000:80'
    environment:
      GITLAB_OMNIBUS_CONFIG: |
        gitlab_rails['initial_root_password'] = 'ajay@123456789'
        puma['worker_processes'] = 0

```



# **✅ 2. Start Docker Compose**
Open PowerShell in the same folder and run:



```
docker compose up

```

# **⚠️ Yes, data will be lost if the container is removed or recreated.**

# **GitLab Server with Docker Volumes for Data Persistence**

**The command docker compose down is used to stop and remove all containers, networks, and volumes (if specified) created by docker compose up.**

**✅Syntax:**



```
docker compose down

```



# **docker-compose.yaml**



```
version: '3.8'

services:
  gitlab-server:
    image: 'gitlab/gitlab-ce'
    container_name: my-gitlab-server
    ports:
      - '3000:80'
    environment:
      GITLAB_OMNIBUS_CONFIG: |
        gitlab_rails['initial_root_password'] = 'ajay@123456789'
        puma['worker_processes'] = 0
    volumes:
      - ./gitlab/config:/etc/gitlab
      - ./gitlab/logs:/var/log/gitlab
      - ./gitlab/data:/var/opt/gitlab

```
