# **What is DevOps?**

**DevOps is a methodology that brings together Development (Dev) and Operations (Ops) teams to improve software delivery speed, quality, and reliability. It focuses on automation, collaboration, and continuous improvement.**

image.png

# **what is mean by Development(Creating Software )**

This part focuses on writing, testing, and improving software.

Who? Developers, testers.

What they do?

 Write & update code.

 Test features & fix bugs.

 Use version control (e.g., Git).

 Example: A developer writes code for a new payment feature in an e-commerce app.

# **what is mean by Operations (Running & Maintaining Software )**

This part ensures that the software runs smoothly after deployment.

Who? IT admins, system engineers.

What they do?

 Deploy & manage applications.

 Monitor servers & fix issues.

 Ensure security & scalability.

 Example: The operations team ensures the payment system is live, secure, and fast, even during high traffic.

# **Why is DevOps Important?**

Faster Releases  Automates processes to deploy software quickly.

Better Quality  Continuous testing ensures fewer bugs.

Improved Collaboration  Developers & IT teams work together.

Scalability  Easily handle growing demands.

**How is DevOps Used in Real Life?**

Continuous Integration & Continuous Deployment (CI/CD)

Automates code testing and deployment.

Example: A banking app gets daily updates without downtime.

Tools: Jenkins, GitLab CI/CD, Azure DevOps.

**Cloud & Containerization**

Deploys applications in scalable cloud environments.

Example: Netflix uses AWS & Kubernetes to stream videos smoothly.

Tools: AWS, Docker, Kubernetes.

**Tracks app performance & prevents cyberattacks.**

Example: E-commerce websites monitor traffic spikes to prevent crashes.

Tools: Prometheus, Grafana, Splunk, SonarQube.

**Real-Life Example**

 E-commerce Website (e.g., Amazon, Flipkart)

Developer pushes new features (e.g., faster checkout)

# **ROADMAP**

**1. Version Control System: Git**

Tracks changes in source code during software development.

Example: Managing branches and commits in a project.

**2. Hosting Platform: GitHub, GitLab**

Hosts Git repositories with features like collaboration, CI/CD, and issue tracking.

Example: GitHub Actions for automated deployments.

**3. Operating System: Linux**

Open-source OS widely used in servers and DevOps environments.

Example: Running Ubuntu on a virtual machine for development.

**4. Docker (Containerization)**

Packs your app and its setup so it runs the same everywhere.

Example: Running a Python app inside a Docker container.

**5. Cloud Provider: AWS, Azure, Google Cloud**

Offers scalable cloud infrastructure and services for hosting and deploying apps.

Example: Deploying a website using AWS EC2.

**6.Configuration Management -Ansible**

Automatically sets up software on many machines.

Example: Installing software on 10 servers with one Ansible command.

**7.Infrastructure Provisioning-Terraform**

Builds servers and resources using code.

Example: Creating a network and server on AWS using a script.

**8. Infrastructure Monitoring: Grafana**

Visualizes system metrics and logs for monitoring infrastructure performance.

Example: Monitoring CPU and memory usage on a server using Grafana dashboards.

**9. Container Orchestration: Kubernetes**

Manages container deployment, scaling, and networking.

Example: Deploying and scaling microservices using Kubernetes clusters.

# ** What is Git?**
Git is a free, open-source version control system that helps you track changes in your code over time. It allows multiple developers to collaborate, manage different versions of code, and safely undo mistakes.


# ** Step-by-step Git Commands to Push Code to GitHub:(Git Bash & VS code)**



```
git init                                     # Initialize Git in your local folder
git add .                                    # Add all files to staging
git commit -m "Initial commit"               # Commit your files with a message
git branch -M main                           # Rename default branch to 'main'
git remote add origin https://github.com/YourUsername/YourRepo.git  # Add remote GitHub repo
git pull --rebase origin main                # Optional: sync if repo is not empty
git push -u origin main                      # Push local code to GitHub

```

# ** Types of Repositories in Git:**
**Local Repository**

This is on your own computer.

You create it using git init.

Example: You're working on your code before sharing it.

**Remote Repository**

This is hosted on platforms like GitHub, GitLab, or Bitbucket.

You connect to it using git remote add origin <URL>.

Used to share your code with others or store it online.
