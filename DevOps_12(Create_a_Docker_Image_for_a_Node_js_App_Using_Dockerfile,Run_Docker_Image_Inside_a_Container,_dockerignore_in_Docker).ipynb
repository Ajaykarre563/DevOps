

# **dockerfile**



```
# Use official Node.js image (you can pin a version like node:18 for stability)
FROM node:18

# Set working directory
WORKDIR /app

# Copy only package files first for layer caching
COPY package*.json ./

# Install dependencies
RUN npm install

# Copy the rest of your app
COPY . .

# Expose the app port
EXPOSE 3000

# Run the app
CMD ["node", "index.js"]

```

# **.dockerignore in Docker**

# ** What is .dockerignore in Docker?**
The .dockerignore file tells Docker which files and folders to ignore when building an image  just like .gitignore in Git.

# ** Why is it Useful?**

Speeds up image build

Reduces image size

Prevents sensitive or unnecessary files from being copied



# ** Example .dockerignore**



```
node_modules
npm-debug.log
Dockerfile
.dockerignore
.git
.gitignore
*.md

```
