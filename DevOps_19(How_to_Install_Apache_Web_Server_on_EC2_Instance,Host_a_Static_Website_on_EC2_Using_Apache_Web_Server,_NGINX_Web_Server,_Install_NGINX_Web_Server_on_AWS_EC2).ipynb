# ** Step 5: Copy Files to Apache Directory**



```
sudo cp -r /tmp/Project/* /var/www/html/

```



# ** What it Means:**
sudo: Run with admin/root privileges

cp: Copy command

-r: Recursive  copies folders and files inside them

/tmp/Project/*: All files and subdirectories from the Project folder located in /tmp

/var/www/html/: Destination directory for Apache web server content
