# **‚úÖ Elastic IP (EIP) in AWS**(IP Address)
An Elastic IP is a static public IPv4 address that you can assign to your EC2 instance.

üîπ Key Points:
Unlike regular public IPs (which change when the instance stops/starts), an Elastic IP does not change.

You can reassign it to different EC2 instances as needed.

Useful for web servers, NAT gateways, or any service where a consistent IP is needed.

# **‚úÖ EBS (Elastic Block Store) in AWS**(Storage)
EBS is AWS‚Äôs block storage used with EC2 instances‚Äîjust like a hard drive.

üîπ Key Points:
You can attach it to an EC2 instance as its root volume or additional storage.

Data on EBS persists even if the instance stops (unlike EC2 instance store).

Supports snapshots (backups), resizing, and encryption.

Comes in different types like gp3, io1, st1, etc., depending on performance needs.
