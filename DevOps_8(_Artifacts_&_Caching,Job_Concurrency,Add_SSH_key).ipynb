# **üîπ What is Caching?**
Caching means saving frequently used files or data (like dependencies, or Next.js build cache) so that they don‚Äôt need to be downloaded or generated again.

Example: Save node_modules or .next/cache folder

Benefit: Faster builds because it skips re-installing or re-building unchanged parts

üîÅ Think of it like storing instant noodles instead of cooking fresh every time!

# **üîπ What are Artifacts?**
Artifacts are the final output files of a build (like compiled code, zip files, or static site folders) that you save and use later.

Example: Save the .next/ folder or out/ folder for deployment

Benefit: You can use the same build output in multiple CI/CD steps without re-building

üì¶ Think of artifacts like a finished lunchbox you carry to work.

# **WorkFlow code for Next.js**



```
name: Build and Upload Next.js App

on:
  push:
    branches:
      - main
  pull_request:

jobs:
  build:
    runs-on: ubuntu-latest

    steps:
      - name: Checkout Repository
        uses: actions/checkout@v4.2.2

      - name: Setup Node JS
        uses: actions/setup-node@v4.2.0
        with:
          node-version: 22.x
          cache: 'npm'

      - name: Install Dependencies
        run: npm install

      - name: Build Next Js App
        run: npm run build

      - name: Upload Build on Artifacts
        uses: actions/upload-artifact@v4.6.0
        with:
          name: nextjs-build
          path: out/

```

# **üí° What Does Job Concurrency Mean in GitHub Actions?**
Job concurrency in GitHub Actions is a way to control how many jobs or workflows run at the same time for a specific task or group.

# **üõë Why Use It?**
It helps to:

Prevent conflicts (e.g., multiple deployments at once)

Save resources by canceling outdated runs (like older commits)

Ensure that only the latest job runs, especially in fast-moving branches like main or during pull requests

# **Job Concurrency workflow**



```
name: Deploy Next.js App

on:
  push:
    branches:
      - main

jobs:
  deploy:
    runs-on: ubuntu-latest

    concurrency:
      group: deploy
      cancel-in-progress: true

    steps:
      - name: Checkout Repository
        uses: actions/checkout@v4.2.2

      - name: Install Dependencies
        run: npm install

      - name: Build Next.JS App
        run: npm run build

      - name: Deploy App
        run: echo "Deploying App...."

```



# **üìå Note on Job Concurrency in GitHub Actions**
When you make changes (e.g., push new commits), GitHub Actions will cancel the previous running job in the same concurrency group and only run the latest one.
